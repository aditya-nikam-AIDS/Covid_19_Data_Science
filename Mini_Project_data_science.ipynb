```python
male_vaccinated=df['Male(Individuals Vaccinated)'].sum()
print("Number of males Successfullly vaccinated ",male_vaccinated)
```

```python
#Number of Females vaccinated
```

```python
female_vaccinated=df['Female(Individuals Vaccinated)'].sum()
print("Number of females Successfullly vaccinated ",female_vaccinated)
```

```python
import matplotlib.pyplot as plt
plt.bar(['Male','Female'],[male_vaccinated,female_vaccinated])
plt.xlabel('Gender')
plt.ylabel('No of Persons')
plt.title("Vaccinated People")
plt.show()
```
